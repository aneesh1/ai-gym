# Chapter 1 Exercises

1. How would you define machine learning?

    * programming computers so that they can learn from data

2. Can you name four types of problems where it shines?

    * problems whose solutions require a lot of hand-tuning or lots of constraints
    * complicated problems for which a "good" solution cannot be found using a traditional approach
    * environments that are constantly changing (fluctuating)
    * ~gaining insights + finding relationships in complex problems and large amounts of data~
    * data mining - to help humans learn
    
3. What is a labeled training set?

    * A labeled training set is a training set whose instances has the "solution" (aka correct classification) 
    
4. What are the two most common supervised tasks?

    * Classification
    * Regression - Predicting a target numeric value
        * i.e. the price of a car from features

5. Can you name four common unsupervised tasks?

    * Clustering
        * i.e. suppose you have a blog and you have a lot of data
        * you may want to detect groups of similar visitors
    * Visualization
        * simply gives a 2D/3D representation of the data you feed it in
    * Dimensionality Reduction
    * ~Anomaly Detection~
        * finding outliers and removing them from the dataset before feeding them to the learning algorithm
    * Association Rule Learning
        * digging into large amounts of data and trying to find interesting relationships between attributes
        
6. What type of ML algorithm would you use to allow a robot to walk in various unknown terrains?

    * Reinforcement learning would be the best choice because that can teach the robot how to observe the environment and then how to act by giving it rewards for walking correctly and penalties for doing the wrong thing. 
7. What type of algorithm would you use to segment your customers into multiple groups?

    * Clustering
    
    * If you don't know how to define your groups --> use clustering (unsupervised learning)
    * If you know what groups you want --> classification algorithm (supervised learning)
8. Would you frame the problem of spam detection as a supervised learning problem or an unsupervised learning problem?

    * ~I would frame the problem of spam detection as an unsupervised learning problem since this would allow the program to better generalize when it sees a new kind of spam email.~
    * Use a supervised learning model -- feed many emails w/ their label
    
    
9. What is an online learning system?

    * Contrary to popular belief, an online learning system is not actually online, but instead is a system that learns when a person incrementally feeds it instances of data. It is important to note that each time the system learns something, the process is fast and cheap so the system can learn on the fly.  
    * Contrast w/ batch learning
10. What is out-of-core learning?

    * When a dataset cannot fit in one machine's main memory, the algorithm loads part of the data, and then runs a training step on that data and continues to do the same until it has run on all of the data. 
  
11. What type of learning algorithm relies on a similarity measure to make predictions?

    * Instance-based learning
12. What is the difference between a model parameter and a learning algorithm's hyperparameter?

    * An algorithm's hyperparameter does not chance during training and is set prior to training. It also determines how much regularization to apply during learning. For instance, a large hyperparameter means a flat model (slope close to 0). Meanwhile, a model parameter simply represents a function that can fit the data and does change during training since you are looking for one that minimizes the cost function.  
    * Model parameters determine what the model will predict given a new instance
    * Model parameter is for the model, hyperparameter for the learning algorithm
13. What do model based learning algorithms search for?

    * Model based learning algorithms are looking for model parameter values that minimize a cost function. Overall, they are looking for a parameter that will generalize well to new instances. 
    
14. Can you name four of the main challenges in ML?

    * Insufficient Quantity of Training Data
    * Nonrepresentative Training Data
    * Poor Quality Data
    * Irrelevant features
    * Overfitting the Training Data
    * Underfitting the Training Data
15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?

    * The model is overfitting the training data. To fix this, we can: 
        1. use regularization by changing the number of model parameters and therefore the degrees 
        2. Feeding better features (feature engineering)
        3. Reducing constraints on the model (reducing the regularization hyperparameter)
    
    * Book's answers:
        * Getting more data
        * simplifying the model (selecting a simpler algo, reducing number of parameters/features, regularizing the model)
        * reducing the noise in the training data
       
16. What is a test set and why would you want to use it?

    * The test is used to estimate generalization error that a model will make on new instances
    
17. What is the purpose of a validation set?

    * A validation set is meant to ensure that a machine learning model's hyperparameters are not just geared toward generating an accurate model for just one dataset. Using the information from a validation set, we can tune hyperparameters accordingly.  
    
18. What can go wrong if you tune hyperparameters using the test set?
    * Tuning hyperparameters using the test set could cause the model to overfit the test set which could cause issues when the model is launched into production. 
    
    
19. What is cross-validation and why would you prefer it to a validation set?
    * ~Cross-validation is when the training set is split into complementary subsets and each model is trained against a different combination of those subsets. You would prefer this because you wouldn't have to "waste" as much training data in validation sets.~
    * Cross validation lets us compare models without the need for a separate validation set. We can thus save training data.
